```python
# Функция для заполнения пропущенных значений
def fill_dept(row):
    
    """
    Функция для заполнения пропущенных значений в столбце 'dept' на основе данных столбца 'salary'
    """
    
    if row['salary'] < 22000:
        return 'sales'
    elif 22000 <= row['salary'] <= 25000:
        return 'purchasing'
    else:
        return 'technology'
```

```python
# Функция для вывода графика количественных признаков
def graph_quant_feature(df, col_name, col_title):
    
    """
    Функция в качестве аргументов принимает датафрейм, название столбца, для которого необходимо вывести график, 
    и общее название для 2-х графиков. Она выводит гистограмму и диаграмму размаха для полученных данных
    """
    
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
    
    bins = int(np.log2(len(df[col_name])) + 1)
    
    sns.histplot(df[col_name], bins=bins, ax=axes[0], palette='deep', edgecolor='black')
    axes[0].set_xlabel(f'Количественный признак для столбца \'{col_name}\'')
    axes[0].set_ylabel('Количество сотрудников')
    axes[0].set_title('Гистограмма')
    axes[0].grid(True)
    
    sns.boxplot(data=df[col_name], orient='h', ax=axes[1], color='orange')
    axes[1].set_xlabel(f'Количественная характеристика для столбца \'{col_name}\'')
    axes[1].set_title('Диаграмма размаха')
    
    plt.suptitle(col_title)
    plt.show()
    
    print(f'Рассмотрим описательную статистику:\n{df[col_name].describe()}')
```

```python
# Функция для вывода графика категориальных признаков
def graph_cat_feature(df, col_name, col_title):
      
    """
    Функция в качестве аргументов принимает датафрейм, название столбца, для которого необходимо вывести график, 
    и название самого графика. Она выводит столбчатый график для полученных данных
    """
      
    plt.figure(figsize=(6, 6))
    sns.countplot(data=df, x=col_name, palette='deep', edgecolor='black')
    plt.title(col_title)
    plt.ylabel('Количество сотрудников')
    plt.xlabel(f'Значение столбца \'{col_name}\'')
    plt.xticks(rotation=0)
    plt.show()
```

```python
# Функция расчёта метрики SMAPE
def SMAPE_metric(y_true, y_pred):
    
    """
    Функция получает в качестве аргументов фактическое значение целевого признака (y_true), 
    предсказанное значение целевого признака (y_pred) и возвращает значение метрики SMAPE
    """
    
    n = len(y_true)
    smape_metrics = (100 / n) * np.sum(np.abs(y_pred - y_true) / ((np.abs(y_true) + np.abs(y_pred)) / 2))
    return smape_metrics
```

```python
# Функция построения столбчатой диаграммы для создания портрета «уволившегося сотрудника»
def percentage_chart(df, column):
    
    """
    Функция принимает датафрейм и название столбца, по значениям которого строится столбчатая диаграмма с отображением 
    процентного соотношения значений quit='yes' относительно общего количества записей в каждой группе.
    """

    grouped_df = df.groupby(column)['quit'].value_counts(normalize=True).unstack(fill_value=0)
    grouped_df['yes'] = grouped_df['yes'] * 100

    plt.figure(figsize=(10, 6))
    ax = sns.barplot(data=grouped_df, x=grouped_df.index, y='yes', palette='deep', edgecolor='black')
    
    plt.title(f'Анализ текучести персонала по столбцу \'{column}\'')
    plt.xlabel(f'Значение столбца {column}')
    plt.ylabel('Доля уволившихся сотрудников')
    plt.ylim(0, 100)
    plt.grid(axis='y', alpha=0.3)
    
    for p in ax.patches:
        height = p.get_height()
        ax.text(p.get_x() + p.get_width()/2., height + 0.5, 
                f'{height:.1f}%', ha="center", va="bottom")
    
    plt.show()
```

```python
# Объединение тестовой выборки
test_combined_df = test_features.merge(test_jsr, on='id', how='left')

# Проверяем правильность преобразований
test_combined_df.head()
```

Из тренировочной выборки удаляем столбец ID и исключаем возникшие дубликаты записей

```python
# Из тренировочной выборки удаляем столбец ID и исключаем возникшие дубликаты записей
tmp_df = train_jsr.drop('id', axis=1).drop_duplicates()

# Объявляем данные тренировочной выборки
X_train_2 = tmp_df.drop('job_satisfaction_rate', axis=1)
y_train_2 = tmp_df['job_satisfaction_rate']

# Объявляем данные тестовой выборки
X_test_2 = test_combined_df.drop(['job_satisfaction_rate', 'id'], axis=1)
y_test_2 = test_combined_df['job_satisfaction_rate'] 
```

#### Подготовка признаков в пайплайне

```python
# Создаём списки с названиями признаков
ohe_columns = ['dept', 'last_year_promo', 'last_year_violations']
ord_columns = ['level', 'workload']
num_columns = ['employment_years', 'supervisor_evaluation', 'salary']
```

```python
# Создаём пайплайн для подготовки признаков из списка ohe_columns: заполнение пропусков и OHE-кодирование
# SimpleImputer + OHE
ohe_pipe = Pipeline(
                [
                    ('simpleImputer_ohe', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                    ('ohe', OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False))
                ]
)

# Создаём пайплайн для подготовки признаков из списка ord_columns: заполнение пропусков и Ordinal-кодирование
# SimpleImputer + OE
ord_pipe = Pipeline(
            [
                ('simpleImputer_before_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent')),
                ('ord', OrdinalEncoder(categories=[
                                                    ['junior', 'middle', 'senior'], 
                                                    ['low', 'medium', 'high'],
                                                  ],
                                       handle_unknown='use_encoded_value', unknown_value=np.nan)),
                ('simpleImputer_after_ord', SimpleImputer(missing_values=np.nan, strategy='most_frequent'))
            ]
)
```

Пайплайны созданы с учётом заполнения пустых значений наиболее часто встречающимися.

```python
# Создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
                            [
                                ('ohe', ohe_pipe, ohe_columns),
                                ('ord', ord_pipe, ord_columns),
                                ('num', StandardScaler(), num_columns)
                            ],
                            remainder='passthrough'
)
```

**Вывод по разделу:**
- Проведён корреляционный анализ всех признаков в тестовой выборке.
- Исследованы сильные взаимосвязи между признаками.
- Разработан единый пайплайн подготовки признаков.

### Обучение моделей

```python
# Создаём итоговый пайплайн: подготовка данных и модель
pipe_final = Pipeline([
                        ('preprocessor', data_preprocessor),
                        ('models', LinearRegression())
                    ])
```

Для решения задачи регрессии будут использованы модели LinearRegression и DecisionTreeRegressor, которые позволят найти оптимальное решение

```python
# Производим настройку параметров и гиперпараметров моделей машинного обучения для решения поставленной задачи
param_grid = [

        {
            'models': [LinearRegression()],
            'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
        },   
    
        {
            'models': [DecisionTreeRegressor(random_state=RANDOM_STATE)],
            'models__max_depth': range(8, 16),
            'models__max_features': range(8, 11),
            'models__min_samples_split': range(3, 15),
            'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
        } 
    
]
```

```python
# Создаем собственную метрику SMAPE 
# (англ. symmetric mean absolute percentage error, «симметричное среднее абсолютное процентное отклонение»)
smape = make_scorer(SMAPE_metric, greater_is_better=False)
```

```python
# Найдем лучшую модель, используя собственную метрику SMAPE
grid_search2 = GridSearchCV(pipe_final, 
                            param_grid,
                            cv=5,
                            scoring=smape,
                            n_jobs=-1) 

grid_search2.fit(X_train_2, y_train_2)
print('Лучшая модель и её параметры:\n', grid_search2.best_estimator_)
```

```python
# Выводим лучшую метрику 2-й модели на тренировочной выборке
print(f'Метрика SMAPE на тренировочной выборке: {round(abs(grid_search2.best_score_), 5)}')
```

```python
# Выводим лучшие параметры 2-й модели
grid_search2.best_params_
```

```python
# Установка опции для отображения максимальной ширины столбца
pd.set_option('display.max_colwidth', None)

# Получаем все результаты и создаем из них датафрейм, отсортируем по рейтингу качества моделей:
result = pd.DataFrame(grid_search2.cv_results_)
result[['rank_test_score', 'mean_test_score','params']].sort_values('rank_test_score').head(10)
```

В топ-10 лучших моделей вошли варианты DecisionTreeRegressor с различными наборами гиперпараметров

```python
# Предсказываем значения на тестовой выборке и выводим метрику SMAPE
y_pred_2 = grid_search2.best_estimator_.predict(X_test_2)
print(f'Метрика SMAPE на тестовой выборке: {round(SMAPE_metric(y_test_2, y_pred_2), 5)}')
```

### Обучение модели

```python
# Создаём общий пайплайн для подготовки данных
data_preprocessor_2 = ColumnTransformer(
                            [
                                ('ohe', ohe_pipe, ohe_columns_2),
                                ('ord', ord_pipe, ord_columns_2),
                                ('num', StandardScaler(), num_columns_2)
                            ],
                            remainder='passthrough'
)
```

```python
# Создаём итоговый пайплайн: подготовка данных и модель
pipe_final2 = Pipeline([
                        ('preprocessor', data_preprocessor_2),
                        ('models', LogisticRegression(random_state=RANDOM_STATE))
                    ])
```

Для решения задачи классификации будет проведен рандомизированный поиск оптимальной модели среди следующих алгоритмов: LogisticRegression, KNeighborsClassifier, DecisionTreeClassifier. Для каждой модели будет выполнена настройка не менее двух ключевых гиперпараметров с целью выявления наилучшей конфигурации

```python
# Производим настройку параметров и гиперпараметров моделей машинного обучения для решения поставленной задачи
param_grid_2 = [

        {
            'models': [KNeighborsClassifier()],
            'models__n_neighbors': range(3,5),
            'models__weights': ['uniform', 'distance'],
            'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
        },    
        
        {
            'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
            'models__max_depth': range(8,10),
            'models__max_features': range(3,7),
            'models__min_samples_split': range(3,7),
            'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
        },

        {
            'models': [LogisticRegression(random_state=RANDOM_STATE, penalty='l1', solver='liblinear')],
            'models__C': range(1,5),
            'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']
        }
    
]
```

```python
# Найдем лучшую модель, используя метрику 'roc_auc'
randomized_search = GridSearchCV(pipe_final2, 
                                       param_grid_2,                                       
                                       cv=5,
                                       scoring='roc_auc',
                                       n_jobs=-1) 

randomized_search.fit(X_train, y_train)
print('Лучшая модель и её параметры:\n', randomized_search.best_estimator_)
```

```python
# Выводим лучшие параметры полученной модели
randomized_search.best_params_
```

```python
# Выводим метрику лучшей модели на тренировочной выборке
print ('Метрика лучшей модели на тренировочной выборке:', round(randomized_search.best_score_, 5))
```

```python
# Получаем все результаты и создаем из них датафрейм, отсортируем по рейтингу качества моделей:
result_2 = pd.DataFrame(randomized_search.cv_results_)
result_2[['rank_test_score', 'mean_test_score','params']].sort_values('rank_test_score').head(10)
```

```python
# Выводим ROC-AUC на тестовой выборке
y_pred_proba = randomized_search.predict_proba(X_test)
print(f'Метрика ROC-AUC на тестовой выборке: {round(roc_auc_score(y_test, y_pred_proba[:, 1]), 5)}')
```

```python
# Получаем доступ к лучшей модели
best_model = randomized_search.best_estimator_['models']

# Получаем препроцессор
preprocessor = randomized_search.best_estimator_['preprocessor']
```

```python
# Строим диаграмму важности признаков
plt.figure(figsize=(10, 6))
sns.barplot(data=importance_df.sort_values(by='Importance', ascending=False), x='Importance', y='Feature',
            palette='deep', edgecolor='black')
plt.title('Важность признаков для DecisionTreeClassifier')
plt.xlabel('Важность признака')
plt.ylabel('Название признака')
plt.show()
```

**Промежуточный вывод:** Модель DecisionTreeClassifier показала лучший результат благодаря способности эффективно выявлять сложные нелинейные зависимости между признаками, а настройка гиперпараметра max_features = 5 помогла отобрать наиболее информативные характеристики, существенно влияющие на целевой признак, что значительно улучшило качество классификации.
