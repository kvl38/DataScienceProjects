## Предобработка данных

Создадим функцию для отображения основных сведений об указанном столбце датафрейма

```python
def column_info(data, col_name):
    
    """
    В качестве аргументов функция принимает датафрейм для анализа (data) и название столбца (col_name)
    Эта функция будет выводить информацию о количестве пропущенных и уникальных значений и о типе данных столбца
    для указанного столбца в датафрейме
    """
    
    # Количество пропущенных значений
    print(f"Количество пропущенных значений: {data[col_name].isna().sum()}")
    
    # Количество уникальных значений
    print(f"Количество уникальных значений: {data[col_name].nunique()}")
    
    # Тип данных
    print(f"Тип данных: {data[col_name].dtype}")
```

```python
# Выделяем отсортированные данные для каждого региона
na_platforms = current_df.groupby('platform')['na_sales'].sum().sort_values(ascending=False)
eu_platforms = current_df.groupby('platform')['eu_sales'].sum().sort_values(ascending=False)
jp_platforms = current_df.groupby('platform')['jp_sales'].sum().sort_values(ascending=False)

# Выводим полученный датасет для проверки его корректности
jp_platforms
```

```python
# Функция для подготовки данных для круговой диаграммы
def prepare_pie_data(data):
        
    """
    В качестве аргумента функция принимает датафрейм для анализа (data)
    Эта функция будет извлекать топ-5 платформ из выбранного датасета, а продажи остальных платформ объединять в категорию "Другие"
    """
    
    top5_df = data.head(5)
    pie_data = top5_df.append(pd.Series({"Другие": data.sum() - top5_df.sum()}))
    return pie_data
```

```python
# Построение всех диаграмм
fig, axs = plt.subplots(1, 3, figsize=(18, 6))
for i, (data, region_name) in enumerate(zip([na_platforms, eu_platforms, jp_platforms], ["Северной Америке (NA)", "Европе (EU)", "Японии (JP)"])):
    pie_data = prepare_pie_data(data)
    pie_data.plot.pie(ax=axs[i], autopct='%1.1f%%', startangle=90)
    axs[i].set_title(f'Топ 5 платформ в {region_name}')
    axs[i].set_ylabel('')

plt.show()
```

В Северной Америке 24.8% рынка занимает PS4, почти столько же продаж игр приходится на Xbox One (21.3%), а треть рынка делят Xbox 360 (18.7%) и PS3 (14.5%). В Европе, аналогично, лидируют игры для PS4, занимая 36% всего рынка, при этом другую треть занимают PS3 (17.3%) и Xbox One (13.2%). Рынок Японии кардинально отличается от Европы и Северной Америки: почти половину рынка занимают игры для 3DS (48.2%), за ними следуют PS3 (16.6%) и PS Vita (13.2%), а PS4 замыкает список лидеров с долей в 11.3% рынка

## Проверка гипотез

### Средние пользовательские рейтинги платформ Xbox One и PC

```python
# Создадим датасет с данными пользовательского рейтинга платформы Xbox One
xone_df = current_df[~(current_df['user_score'].isna()) & (current_df['platform'] == 'xone')]['user_score']

# Изучим статистическую сводку о данных пользовательского рейтинга платформы Xbox One
xone_df.describe()
```

```python
# Создадим датасет с данными пользовательского рейтинга платформы PC
pc_df = current_df[~(current_df['user_score'].isna()) & (current_df['platform'] == 'pc')]['user_score']

# Изучим статистическую сводку о данных пользовательского рейтинга платформы PC
pc_df.describe()
```

Проводится исследование выборок пользовательских оценок игр за актуальный период для платформ Xbox One и PC. 

* Нулевая гипотеза (H0): Средние пользовательские рейтинги платформ Xbox One и PC равны.

* Альтернативная гипотеза (H1): Средние пользовательские рейтинги платформ Xbox One и PC не равны.

Для анализа применяется двухвыборочный t-тест для независимых выборок. Задаем уровень значимости alpha = 0.05.

```python
# Задаем уровень значимости
alpha = 0.05

# Проверка гипотезы
platform_result = st.ttest_ind(xone_df, pc_df, equal_var=False)

# Выводим на экран значение p_value
print('p_value:' , platform_result.pvalue)

# Выводим на экран результат
if platform_result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

*Вывод:* Были рассмотрены две выборки, объемы данных в которых незначительно отличаются. Анализ описательной статистики и результаты проверки гипотез показали, что средние пользовательские рейтинги платформ Xbox One и PC не имеют значимых различий.

### Средние пользовательские рейтинги жанров Action и Sports

```python
# Создадим датасет с данными пользовательского рейтинга игр жанра Action
action_df = current_df[~(current_df['user_score'].isna()) & (current_df['genre'] == 'action')]['user_score']

# Изучим статистическую сводку о данных пользовательского рейтинга игр жанра Action
action_df.describe()
```

```python
# Создадим датасет с данными пользовательского рейтинга игр жанра Sports
sports_df = current_df[~(current_df['user_score'].isna()) & (current_df['genre'] == 'sports')]['user_score']

# Изучим статистическую сводку о данных пользовательского рейтинга игр жанра Sports
sports_df.describe()
```

Проводится исследование выборок пользовательских оценок игр за актуальный период для игр жанра Action и Sports 

* Нулевая гипотеза (H0): Средние пользовательские рейтинги игр жанров Action и Sports равны.

* Альтернативная гипотеза (H1): Средние пользовательские рейтинги игр жанров Action и Sports не равны.

Для анализа применяется двухвыборочный t-тест для независимых выборок. Задаем уровень значимости alpha = 0.05.

```python
# Задаем уровень значимости
alpha = 0.05

# Проверка гипотезы
genre_result = st.ttest_ind(action_df, sports_df, equal_var=False)

# Выводим на экран значение p_value
print('p_value:' , genre_result.pvalue)

# Выводим на экран результат
if genre_result.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

*Вывод:* Были рассмотрены две выборки, объемы данных в которых различаются: данных о пользовательских оценках игр жанра Action в 2,3 раза больше, чем данных о пользовательских оценках игр жанра Sports. Анализ описательной статистики и результаты проверки гипотез показали, что средние пользовательские рейтинги игр жанров Action и Sports существенно различаются.

### Пояснение действий

#### Формулировка нулевой и альтернативной гипотез

Нулевая гипотеза, как правило, предполагает равенство средних значений сравниваемых выборок (отсутствие значимой разницы), в то время как альтернативная гипотеза утверждает, что средние значения выборок различаются (существует значимая разница). В рамках проведенных исследований отсутствует указание на возможное направление различий средних значений в случае их неравенства (какое из них больше — первое или второе). Поэтому рассматриваются только два случая: либо средние значения выборок равны, либо они различаются.

#### Критерии для проверки гипотез

В ходе исследования применялся t-тест Стьюдента. Уровень значимости был установлен равным 0,05, что является часто используемым значением для проверки гипотез. Также использовался параметр equal_var=False, поскольку равенство дисперсий рассматриваемых выборок было неизвестно. Параметр alternative='two-sided', который указывает на использование двусторонней альтернативной гипотезы, не указывался, так как он является значением по умолчанию.
