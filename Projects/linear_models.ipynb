## Моделирование удоев и качества молока для хозяйства «Вольный луг»

**Описание проекта:**

Фермер, владелец молочного хозяйства «Вольный луг», заключил контракт с ассоциацией пастбищ «ЭкоФерма» для расширения поголовья своего стада коров. Ему необходимо тщательно отбирать коров, чтобы каждая бурёнка давала не менее 6000 килограммов молока в год, и при этом молоко соответствовало его строгим критериям вкуса.

**Цель проекта:**

Разработка двух прогнозных моделей машинного обучения для фермера, которые помогут объективно оценивать и отбирать коров: одна для прогнозирования удоя, другая для оценки вероятности получения вкусного молока.

**Задачи:**

- Сбор данных о коровах от ассоциации «ЭкоФерма»
- Разработка модели для прогнозирования возможного удоя коровы
- Создание модели для оценки вероятности получения вкусного молока
- Отбор коров по критериям: средний удой за год ≥ 6000 кг и вкусное молоко
- Предоставление фермеру данных для принятия решений о покупке коров.

### 3 Исследовательский анализ данных

Для удобства создадим две функции: одна будет выводить графики для количественных признаков, а другая — для категориальных

```python
# Функция для вывода графика количественных признаков
def graph_quant_feature(df, col_name, col_title):
    
    """
    Функция в качестве аргументов принимает датафрейм, название столбца, для которого необходимо вывести график, 
    и общее название для 2-х графиков. Она выводит гистограмму и диаграмму размаха для полученных данных
    """
    
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12,6))
    
    axes[0].hist(df[col_name], bins=20)
    axes[0].set_xlabel(f'Количественный признак для столбца \'{col_name}\'')
    axes[0].set_ylabel('Количество коров')
    axes[0].set_title('Гистограмма')
    
    axes[1].boxplot(df[col_name], vert=False)
    axes[1].set_xlabel(f'Количественная характеристика для столбца \'{col_name}\'')
    axes[1].set_title('Диаграмма размаха')
    
    plt.suptitle(col_title)
    plt.show()
```

```python
# Функция для вывода графика категориальных признаков
def graph_cat_feature(df, col_name, col_title):
      
    """
    Функция в качестве аргументов принимает датафрейм, название столбца, для которого необходимо вывести график, 
    и название самого графика. Она выводит столбчатый график для полученных данных
    """
    
    grouped_df = df.groupby(col_name)[col_name].count()
    grouped_df.plot(kind='bar', figsize=(6, 6), grid=True)
  
    plt.title(col_title)
    plt.xlabel(f'Категориальная характеристика для столбца \'{col_name}\'')
    plt.ylabel('Количество коров')
    
    plt.xticks(rotation=0)
 
    plt.show()
```

### 5 Обучение модели линейной регрессии

```python
# Создадим функцию для расчета метрик MSE, MAE, RMSE
def metrics_сalculating(test_value, pred):
    
    mse = mean_squared_error(test_value, pred)
    #rmse = mean_squared_error(test_value, pred, squared=False)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(test_value, pred)
    
    return mse, rmse, mae
```

```python
# Создадим функцию для анализа остатков
def residual_analysis(test_value, pred):
    residuals = test_value - pred

    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(16, 8))

    axes[0].hist(residuals)
    axes[0].set_title('Гистограмма распределения остатков')
    axes[0].set_xlabel('Остатки')
    axes[0].set_ylabel('Количество значений')

    axes[1].scatter(y=pred, x=residuals)
    axes[1].set_xlabel('Остатки')
    axes[1].set_ylabel('Предсказания модели')
    axes[1].set_title('Анализ дисперсии')

    plt.show()
```

### 6 Обучение модели логистической регрессии

```python
# Изменяем значения столбца 'taste_milk' на 1, если оно имеет значение "вкусно", и на 0, если оно "не вкусно"
ferma_main_df['taste_milk'] = ferma_main_df['taste_milk'].apply(lambda x: 1 if x == 'вкусно' else 0)

# Проверяем преобразования
ferma_main_df.head()
```

```python
# Готовим данные для обучения и тестирования модели
X = ferma_main_df.drop(['id', 'taste_milk', 'square_eke'], axis=1)
y = ferma_main_df['taste_milk']

RANDOM_STATE = 42
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=RANDOM_STATE)

cat_col_names = ['breed', 'pasture_type', 'dad_breed', 'spo', 'age']
num_col_names = ['eke', 'crude_protein', 'fat_content', 'protein']
```

```python
# Подготовка признаков
encoder = OneHotEncoder(drop='first', handle_unknown='ignore', sparse_output=False)
X_train_ohe = encoder.fit_transform(X_train[cat_col_names])
X_test_ohe = encoder.transform(X_test[cat_col_names])
encoder_col_names = encoder.get_feature_names_out()

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train[num_col_names])
X_test_scaled = scaler.transform(X_test[num_col_names])

X_train_ohe = pd.DataFrame(X_train_ohe, columns=encoder_col_names)
X_test_ohe = pd.DataFrame(X_test_ohe, columns=encoder_col_names)
X_train_scaled = pd.DataFrame(X_train_scaled, columns=num_col_names)
X_test_scaled = pd.DataFrame(X_test_scaled, columns=num_col_names)

X_train = pd.concat([X_train_ohe, X_train_scaled], axis=1)
X_test = pd.concat([X_test_ohe, X_test_scaled], axis=1)
```

```python
# Обучение модели логистической регрессии на подготовленных данных
clf = LogisticRegression()
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)
y_pred
```

```python
# Оценка качества модели на тестовой выборке
acc = round(accuracy_score(y_test, y_pred), 3)
recall = round(recall_score(y_test, y_pred), 3)
precision = round(precision_score(y_test, y_pred), 3)

# Выводим значения метрик
print('Accuracy =', acc)
print('Recall =', recall)
print('Precision =', precision)
```

```python
# Выводим матрицу ошибок
plt.figure(figsize = (10, 5))
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.title('Матрица ошибок')
plt.ylabel('Настоящие значения')
plt.xlabel('Предсказанные значения')
```

Необходимо сократить число ложно положительных предсказаний (ошибок первого рода), то есть свести число 46 к нулю, вследствие чего повысить метрику Precision до 1. Изменим порог принадлежности к классам

```python
# Получаем вероятности принадлежности каждого объекта в тестовом наборе данных к каждому из классов
y_proba = clf.predict_proba(X_test)[:, 1]

# Создаем и выводим датафрейм с истинными метками класса и их предсказанной вероятностью
data = pd.DataFrame(zip(y_test, y_proba), columns = ['y_valid', 'y_proba']).sort_values(by='y_proba', ascending=False)
data
```

```python
# Созданем массив из 60 равномерно распределенных чисел
thresholds = [round(i, 2) for i in np.linspace(0.4, 1, num = 60, endpoint=False)]

# Создаем столбцы с предсказаниями на основе текущего порога
for i in thresholds:
    data['y_pred_' + str(i)] = data['y_proba'].apply(lambda x: 1 if x >= i else 0)
    
# Выводим 5 случайных строк
data.sample(5)
```

```python
# Вычисляем Precision для текущего порога
threshold_value = 0

for i in thresholds:
    y_pred = data['y_pred_' + str(i)]
    precision = precision_score(y_test, y_pred)
    
    # Выводим текущий порог и значение Precision
    print(f'Threshold: {i}\t Precision: {round(precision, 3)}')

    # Если Precision достигла 1, выводим найденный порог и завершаем цикл
    if precision == 1:
        threshold_value = i
        print(f'\nНеобходимый порог равен: {i}')
        break
```

```python
# Оценка качества модели на тестовой выборке
acc_2 = round(accuracy_score(y_test, data['y_pred_'+str(threshold_value)]), 3)
recall_2 = round(recall_score(y_test, data['y_pred_'+str(threshold_value)]), 3)
precision_2 = round(precision_score(y_test, data['y_pred_'+str(threshold_value)]), 3)

# Выводим значения метрик
print('Accuracy =', acc_2)
print('Recall =', recall_2)
print('Precision =', precision_2)
```

```python
# Выводим матрицу ошибок
plt.figure(figsize = (10, 5))
cm = confusion_matrix(y_test, data['y_pred_'+str(threshold_value)])
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues_r')
plt.title('Матрица ошибок')
plt.ylabel('Настоящие значения')
plt.xlabel('Предсказанные значения')
```
